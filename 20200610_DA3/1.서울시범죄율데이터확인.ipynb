```python
import numpy as np
import pandas as pd
from sklearn import preprocessing

col = ['강간','강도','살인','절도','폭력']

crime_anal = pd.read_csv("crime_anal.csv", encoding = 'utf-8')
x = crime_anal[col].values
min_max_scaler = preprocessing.MinMaxScaler() # 뭐하는앨까?

x_scaled = min_max_scaler.fit_transform(x.astype(float)) #float 값으로 바꿔줌
crime_anal_norm = pd.DataFrame(x_scaled,    columns = col, index = crime_anal.index) #??

col2 = ['강간검거율','강도검거율','살인검거율','절도검거율','폭력검거율']
crime_anal_norm[col2] = crime_anal[col2]
crime_anal_norm.head()
```

```python
result_CCTV = pd.read_csv('../20200608_pyda1/01. CCTV_result.csv', encoding='UTF-8')
result_CCTV
crime_anal_norm[['인구수','CCTV']] = result_CCTV[['인구수','소계']]
crime_anal_norm.head()
```

```python
col = ['강간','강도','살인','절도','폭력']
crime_anal_norm['범죄'] = np.sum(crime_anal_norm[col], axis=1)
crime_anal_norm
```
