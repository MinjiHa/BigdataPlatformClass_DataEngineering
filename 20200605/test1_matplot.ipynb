```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

x = np.arange(0, math.pi * 2, 0.05)
x
```

```python
y = np.sin(x)
plt.plot(x, y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show
```

```python
from pylab import *
x = linspace(-3,3,30) #정렬해서 보여줌
x
y = x**2
y1 = x**4
plot(x,y,'r')
plot(x,y,'r.')
plot(x,y1,'b')
plot(x,y1+10,'m')
#symbols : -,+.....
#b,g,r,c,m,y,k,w
show()
```

```python
plot(x, sin(x))
plot(x,cos(x),'r*')
plot(x,-sin(x),'g--')
plt.xlabel("x axis")
plt.ylabel("y axis")
show()
```

```python
x1 = np.arange(0,math.pi*2,0.05)
y1 = np.sin(x1)
fig = plt.figure()
ax = fig.add_axes([0,0,0.5,1]) #left, bottom, width, height 사이즈조절
ax.plot(x1,y1)
ax.legend(labels=('Smart Phone'), loc = 'lower left') #그래프라벨링추가 / loc설정 없으면 기본값으로
ax.set_title('Sine graph') #그래프타이틀
ax.set_xlabel("x axis") # x축 라벨
ax.set_ylabel("y axis") # y축 라벨
plt.show()
```
