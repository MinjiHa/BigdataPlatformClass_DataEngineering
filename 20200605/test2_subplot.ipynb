```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
fig = plt.figure()
x = np.arange(1,5)
a = fig.subplots(2,2)

a[0][0].grid(True) #grid
a[0][0].set_title("squere") #title
a[0][0].plot(x,x*x,'red',lw=10) #draw chart
a[0][1].plot(x,np.sqrt(x))
a[0][1].grid(color = 'blue', ls='-.', lw=3) # ls 선 종류, lw 선 두께
a[1][0].plot(x,np.exp(x))
a[1][1].plot(x,np.log10(x))
#plt.show()
```

```python
fig = plt.figure(figsize=(4,4)) #인치단위 설정
a1 = fig.add_axes([0,0,1,1])
x4 = np.arange(1,10)
a1.plot(x4, np.exp(x4))
a1.set_title('exp')
a1.set_ylim(0,10000) # y축 늘리기
a1.set_xlim(0,10) # x축 늘리기
```

```python
x5 = np.arange(1,10)
y5 = np.sin(x5)
plt.plot(x5,y5,'b')
plt.xlim(0,10)
plt.ylim(-1.5,1.5)
plt.xticks([0,5,10]) #x축 눈금 수정
plt.yticks([-1,0,1]) #y축 눈금 수정
```

```python
x5 = np.arange(0, math.pi*2, 0.05)
fig = plt.figure(figsize=(8,5))
a5 = fig.add_axes([0.1,0.1,0.8,0.8])
a5.plot(x5, np.exp(x5))
#a5.set_ylim(0,500)
a5.set_ylabel('exp')
a6 = a5.twinx() #축이 하나 더 생김?
a6.plot(x5, np.log(x5),'rx')
a6.set_ylabel('log')
fig.legend(labels=('exp','log'), loc='best')
```

```python
# Bar plot
fig = plt.figure(figsize=(9,4))
languages = ['java', 'javascript','python','html5','css3','sql','c++','c']
students = [29,15,20,3,2,10,15,20]
students2 = [25,10,5,20,2,10,19,10]
plt.bar(languages, students, color='g')
plt.bar(languages, students2, color='y') #뒤에 한게 위에 얹어지는듯
plt.show()
```

```python
from numpy.random import randn

plt.plot(randn(30).cumsum(),'ko--')
plt.plot(randn(30).cumsum(),'k-', drawstyle='steps-post')
```

```python
# 파이썬 차트의 인터페이스가 안예쁨.seaborn 라이브러리 추가해서 쓰면 예쁨.
#2차원 데이터
data = [[30,25,50,20],[40,23,51,17],[35,22,45,19]]
x6 = np.arange(4)
fig = plt.figure(figsize=(8,4))
ax6 = fig.add_axes([0,0,1,1])

ax6.bar(x6 + 0.00, data[0], color='c', width=0.25)
ax6.bar(x6 + 0.25, data[1], color='lightblue', width=0.25)
ax6.bar(x6 + 0.50, data[2], color='gray', width=0.25)
fig.legend(labels=('developer','dba','system engineer'), loc="best")
```

```python
n = 5
menMeans = (20,35,30,35,27)
womenMeans = (25,32,20,19,35)
ind = np.arange(n)
plt.bar(ind, menMeans, width=0.35, color='c')
plt.bar(ind, womenMeans, width=0.35, bottom = menMeans, color='lightblue') #밑에들어갈 바의 값을 bottom = 에 넣어주면 안겹침.
plt.xticks(ind, ('G1','G2','G3','G4','G5'))
plt.ylabel('Scores')
plt.yticks(np.arange(0,81,10))
plt.legend(labels=['Men','Women'])
```
