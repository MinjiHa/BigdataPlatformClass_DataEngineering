```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

x = np.arange(0, math.pi * 2, 0.05)
x
```

```python
y = np.sin(x)
plt.plot(x, y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show
```

```python
x3 = np.arange(0, math.pi * 2, 0.05)
y3 = sin(x3)
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.55,0.55,0.3,0.3])
axes1.plot(x3, y3, 'b')
axes2.plot(x3, np.cos(y3), 'r')
axes1.set_title('sine')
axes2.set_title('cosine')
plt.show()
```
