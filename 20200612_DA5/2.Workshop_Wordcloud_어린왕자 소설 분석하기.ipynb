```python
from konlpy.tag import Twitter
from wordcloud import WordCloud, STOPWORDS
import numpy as np

t = Twitter()
```

```python
# readline_test.py
f = open("./어린왕자.txt", 'r', encoding = 'cp949')
data = f.read()


print(data)
f.close()
```

```python
t.pos(data)
```

```python
import matplotlib.pyplot as plt
import platform
from matplotlib import font_manager, rc
from PIL import Image

# 한글 폰트 설정
rc('font', family='NanumGothic')
%matplotlib inline
```

```python
# 이미지 로드
little_prince = np.array(Image.open('./littleprince.jpg'))

# 검색 제외 단어 설정
stopwords = set(STOPWORDS)
stopwords.add('.')

plt.figure(figsize=(8,8))
plt.imshow(little_prince, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color = 'white', max_words=2000, mask =  little_prince, stopwords = stopwords)
wc = wc.generate(data)
wc.words_
```
