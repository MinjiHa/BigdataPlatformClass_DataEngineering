```python
import pandas as pd

df = pd.read_csv("경찰청_졸음운전교통사고_2014_2016.csv", encoding='cp949')
df.head()
df.info()
df['구분'] = df['구분'].astype('str') #안되네;
df['년도'] = df['구분'].str[:4]
df.head()
```

```python
df2 = df.groupby(by='년도').sum()
#df2['합계'] = df.iloc[:,[1,2,3]].sum(axis=1)
df.info()
df2.head()
```
